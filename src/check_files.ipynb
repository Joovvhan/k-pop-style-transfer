```python
import os
import librosa
from params import params
import scipy.io.wavfile as wavfile
import numpy as np
```

```python
from glob import glob
```

```python
import matplotlib.pyplot as plt
```

```python
path_ori = os.path.join(params.data_dir, params.ori_dir, '*.wav')
path_bit = os.path.join(params.data_dir, params.bit_dir, '*.wav')
path_pno = os.path.join(params.data_dir, params.pno_dir, '*.wav')
```

```python
files_ori = glob(path_ori)
files_bit = glob(path_bit)
files_pno = glob(path_pno)
```

```python
files = [{params.ori_key:ori, params.bit_key:bit, params.pno_key:pno} for ori, bit, pno in zip(files_ori, files_bit, files_pno)]
```

```python
for i in range(1):#range(len(files)):
    y, fs = librosa.core.load(files[i][params.pno_key])
```

```python
plt.plot(y)
plt.show()
```

```python
nsc = int(fs/20)
hop =fs - nsc

# mel = librosa.feature.melspectrogram(y=y, sr=fs, S=None, n_fft=int(fs/100), hop_length=int(fs/100), power=2.0)
Sxx = librosa.core.stft(y, n_fft=nsc, hop_length=hop)
S = abs(Sxx)
log_mel = 20 * np.log10(np.maximum(S, 1e-5))
```

```python
plt.imshow(log_mel, origin='lower', aspect='auto')
plt.show()
```
